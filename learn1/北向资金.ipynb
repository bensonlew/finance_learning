```python
# 北向资金
import pandas as pd
import matplotlib.pyplot as plt
import akshare as ak
import datetime
```

```python
stock_north_flow_in = ak.stock_hsgt_north_net_flow_in_em(symbol="北上")
```

```python
stock_north_all = ak.stock_hsgt_north_acc_flow_in_em("北上")
```

```python
stock_north_all["flow_in"] = stock_north_flow_in["value"]
```

```python
stock_north_flow_in
```

```python
stock_north_all["day"] = stock_north_all["date"].map(str)
stock_north_all.set_index("day", inplace=True)
```

```python
sh000300 = ak.stock_zh_index_daily(symbol="sh000300")
```

```python
sh000300["day"] = sh000300["date"].map(str)
sh000300.set_index("day", inplace=True)
```

```python
sh000300
```

```python
sh300_north = sh000300.merge(stock_north_all, left_index=True, right_index=True, how="left")
```

```python
merge=pd.merge(sh000300, stock_north_all, how='left', left_index=True, right_index=True)
```

```python
merge
```

```python
choose_data = merge[merge.index>"2014-11-17"]
```

```python
choose_data
```

```python
choose_data["flow_in_pct"] = choose_data["flow_in"]/choose_data["value"]
```

```python
choose_data[1000:]
```

```python
choose_data["300_pct"] = choose_data["close"].pct_change()
```

```python
choose_data.to_csv("300_pct.tsv", sep="\t")
```

```python
choose_data["300_prod"] = (choose_data["300_pct"] +1).cumprod()
```

```python
choose_data2 = choose_data[choose_data.index>"2016-01-01"]
```

```python
choose_data2["flow_in_pct_5"] = choose_data2["flow_in_pct"].rolling(5, min_periods=1).mean()
choose_data2["flow_in_pct_20"] = choose_data2["flow_in_pct"].rolling(20, min_periods=1).mean()
choose_data2["flow_in_pct_60"] = choose_data2["flow_in_pct"].rolling(60, min_periods=1).mean()
choose_data2["flow_in_pct_5_diff"] = choose_data2["flow_in_pct_5"] - choose_data2["flow_in_pct_60"]
choose_data2["flow_in_pct_20_diff"] = choose_data2["flow_in_pct_20"] - choose_data2["flow_in_pct_60"]
choose_data2["flow_in_20"] = choose_data2["flow_in"].rolling(20, min_periods=1).mean()/100000
choose_data2["flow_in_60"] = choose_data2["flow_in"].rolling(60, min_periods=1).mean()/100000
```

```python
choose_data2_choose = choose_data2[choose_data2["flow_in_pct_20_diff"] > 0]
choose_data2_choose5 = choose_data2[choose_data2["flow_in_pct_5_diff"]>0]
```

```python
choose_data2["20_60_pct_prod"] = (choose_data2_choose["300_pct"] +1).cumprod()
choose_data2["5_20_pct_prod"] = (choose_data2_choose5["300_pct"] +1).cumprod()
```

```python
choose_data2_choose.to_csv("choose_choose_pct.tsv", sep="\t")
```

```python
plt.rcParams["figure.figsize"] = (20,8)
plt.plot(choose_data2["date_x"], choose_data2["300_prod"], label="300")
plt.plot(choose_data2["date_x"], choose_data2["5_20_pct_prod"].fillna(method="ffill"), label="北向5-20")
plt.legend(loc="best")
```

```python
choose_data2_choose2022 = choose_data2[choose_data2.index>"2021-01-01"]
plt.plot(choose_data2_choose2022["date_x"], choose_data2_choose2022["flow_in_pct_20"])
plt.plot(choose_data2_choose2022["date_x"], choose_data2_choose2022["flow_in_pct_5"])
plt.plot(choose_data2_choose2022["date_x"], choose_data2_choose2022["close"] /400000)
plt.plot(choose_data2_choose2022["date_x"], choose_data2_choose2022["5_20_pct_prod"].fillna(method="ffill") /400)
```

```python
choose_data2_choose2022.to_csv("2022_beixiang.tsv", sep="\t")
```
