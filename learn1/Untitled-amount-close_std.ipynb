```python
import json
import pandas as pd
from matplotlib import pyplot as plt
import glob
import numpy as np
```

```python
files = glob.glob("/mnt/lustre/users/sanger-dev/sg-users/liubinxu/soft/finiance/stock_data/finiance/train/amount_normalize20_rolling_96_exp_mean_close480_close2880_down_K72_9_480_3_150_tree_tree_*_1_target_close1.json")
data_list = []
for n in files:
    amount_pct = float(n.split("tree_tree_")[1].split("_")[0])
    close_std_pct = float(n.split("tree_tree_")[1].split("_")[1])
    f = open(n, "r")
    aa = json.load(f)
    ab = aa.copy()
    ab["choose_k"] = n
    ab["amount_pct"] = amount_pct
    ab["close_std_pct"] = close_std_pct
    data_list.append(ab)
    f.close()
    
```

```python
data_list
```

```python
dfs = pd.DataFrame(data_list)
dfs
```

```python
x=dfs["amount_pct"]
y=dfs["test_day_get30"]



# Fit a polynomial of degree 2 (a quadratic curve)
coefficients = np.polyfit(x, y, 2)
polynomial = np.poly1d(coefficients)

# Generate y-values for the polynomial on a set of x-values
x_values = np.linspace(min(x), max(x), 100)
y_values = polynomial(x_values)



# Show the plot
plt.show()

plt.scatter(x, y)
plt.plot(x_values, y_values, color='r')
```
