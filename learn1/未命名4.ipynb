```python
import numpy as np
import matplotlib.pyplot as plt

# 准备时间序列数据
time = np.linspace(0, 10, num=1000)  # 时间间隔为0.01
# signal = np.sin(2 * np.pi * 5 * time) + 0.5 * np.sin(2 * np.pi * 20 * time)

signal = np.random.rand(1000) + np.sin(2 * np.pi * 20 * time)
# 傅里叶变换
fft_result = np.fft.fft(signal)

# 计算频谱
spectrum = np.abs(fft_result) ** 2

# 可视化频谱
plt.figure(figsize=(8, 4))
freq = np.fft.fftfreq(len(time), d=0.01)  # 计算频率
plt.plot(freq, spectrum)
plt.xlabel('Frequency')
plt.ylabel('Power')
plt.grid(True)
plt.show()
```
